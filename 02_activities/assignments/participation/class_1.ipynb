```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn
```

```python
#Set seed
np.random.seed(613)
# Generate data
x = np.arange(50) # 0,1,2,3,4,...,49
y = np.random.randint(0,100,50) # min, max, number of observations
```

```python
# Scatterplot
fig, ax = plt.subplots(figsize=(5,3))  # Create a figure with axis t width=5 in x height=3 in
ax.scatter(x,y) # Create the scatter plot
```

```python
# Barplot
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.bar(x,y) #bar plot/chart 
```

```python
# Line plot
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.plot(x,y) #create line plot
```

```python
# Histogram
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.hist(y) #Create Historgram plot
```

```python
fig, ax = plt.subplots(figsize=(5,3)) # # Create figure with axis t width=5 in x height=3 in
ax.plot(x,y) #Creae line plot
ax.set_title("Total growth over time") # Title
ax.set_ylabel("Total growth") # Y axis label
ax.set_xlabel("Years since start") # X axis labe
```

```python
# Adjusting dot type 
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.scatter(x,y,    #Creaet scatterokit
           marker ="*", # marker "o"/ "+"
           color = "indigo") #marker color indigo
```

```python
# Line style 
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.plot(x,y,       #Create line plot
         marker ="*", # "o"/ "+"
         color = "indigo", #marker color
            linestyle = "--", #line style
            linewidth = 2)   #line width
```

```python
# Marker color 
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.plot(x,y, 
        marker ="*",  #marker style
        color = "indigo", #line color 
        linestyle = "--", #line sytle
        linewidth = 2, markeredgecolor = "#FA9359", #marker edge color 
         markerfacecolor = "#000000") #marker inside color
```

```python
# Grid line
fig, ax = plt.subplots(figsize=(5,3)) # Create figure with axis t width=5 in x height=3 in
ax.plot(x,y,
          marker ="*",  #Marker
          color = "indigo", #marker colour
          linestyle = "--", #line sytle
          linewidth = 2, #line width
           markeredgecolor = "#FA9359", 
           markerfacecolor = "#000000") 
ax.grid() # both vertical and horizontal grid lines
#ax.grid(axis = "x") # vertical grid lines
#ax.grid(axis = "y") # horizontal grid lines
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Create the plot with styling
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y,
        color='#2E86C1',           # blue color
        marker='o',                # Circle markers
        markersize=6,              # Size of markers
        markerfacecolor='#E74C3C', # Red fill for markers
        markeredgecolor='black',   # Black edge for markers
        markeredgewidth=1,         # Width of marker edge
        linestyle='-',             # Solid line
        linewidth=2,               # Thicker line
        label='Sine Wave')         # Add label for legend

# Customize grid
ax.grid(True, linestyle='--', alpha=0.7)

# Add labels with custom colors
ax.set_xlabel('X axis', color='#2C3E50')
ax.set_ylabel('Y axis', color='#2C3E50')
ax.set_title('Plot with Markers', color='#34495E', fontsize=12)

# Add legend
ax.legend()

# Style the plot background
ax.set_facecolor('#F8F9F9')       # Light gray background
fig.patch.set_facecolor('white')   # White figure background

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
